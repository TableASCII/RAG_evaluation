```python
splitter = TextSplitter()
```

```python
def calc_groundedness(records:list[RAGRecord]):
  answers_sentences = [RAGRecord.answer for RAGRecord in records]
  contexts_sentences = [
        " ".join([doc.text for doc in record.retrieved_docs])
        for record in records
    ]
  results = list(zip(answers_sentences, contexts_sentences))

  groundedness = []

  for answer_text, context_text in results:
    answer_sent = splitter.split_text(answer_text)
    context_sent = splitter.split_text(context_text)

    confirmed_claim = 0

    for answer in answer_sent:
      for context in context_sent:
        sims = cosine_similarity(answer, context)
        if sims > 0.5:
          confirmed_claim += 1
          break
    groundedness.append(confirmed_claim / len(answer_sent))
  return groundedness
```
