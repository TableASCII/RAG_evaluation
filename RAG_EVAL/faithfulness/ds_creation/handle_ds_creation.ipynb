```python
import os

# Список ваших файлов
files_to_combine = ['long_hard_zai_ds_100.jsonl', 'short_simple_gemini_ds_150.jsonl']
output_file = 'bue.jsonl'

def combine_jsonl(files, output_path):
    count = 0
    with open(output_path, 'w', encoding='utf-8') as outfile:
        for filename in files:
            if os.path.exists(filename):
                with open(filename, 'r', encoding='utf-8') as infile:
                    for line in infile:
                        line = line.strip()
                        if line: # Пропускаем пустые строки
                            outfile.write(line + '\n')
                            count += 1
                print(f"Файл {filename} успешно добавлен.")
            else:
                print(f"Внимание: Файл {filename} не найден!")

    print(f"---")
    print(f"Объединение завершено. Всего собрано {count} примеров.")
    print(f"Итоговый файл: {output_path}")

combine_jsonl(files_to_combine, output_file)
```
